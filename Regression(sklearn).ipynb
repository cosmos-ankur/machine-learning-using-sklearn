# What is Linear Regression ?
linear regression is one of the simplest and basic and first step in the world of machine learning
linear regression explores linear relationships between dependent and independent variables


The term “linearity” in algebra refers to a linear relationship between two or more variables. If we draw this relationship in a two-dimensional space (between two variables), we get a straight line.


Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x). So, this regression technique finds out a linear relationship between x (input) and y(output). Hence, the name is Linear Regression.

# working of linear regression
The equation of the above line is :


Y= mx + b


Where b is the intercept and m is the slope of the line. So basically, the linear regression algorithm gives us the most optimal value for the intercept and the slope (in two dimensions). The y and x variables remain the same, since they are the data features and cannot be changed. The values that we can control are the intercept(b) and slope(m). There can be multiple straight lines depending upon the values of intercept and slope. Basically what the linear regression algorithm does is it fits multiple lines on the data points and returns the line that results in the least error.

```python
#creating a dataset
#we will create a artificial dataset 
x = np.linspace(0,100,100000) + np.random.normal(-5,5,100000) #independent variable with some random noise
y = np.linspace(0,100,100000) + np.random.normal(-4,4,100000) #dependent variable with some random noise
```

```python
y
```

```python
#to visualize that our data is linear
plt.plot(x,y,'r')
```

```python
#now to exploit our data we would convert it into a pandas dataframe
feature = pd.DataFrame(x,columns=['Feature'])
target = pd.DataFrame(y,columns=['Target'])
```

```python
#so now we would combine both of them to create one pandas data_set
my_data = pd.concat([feature,target],axis = 1)
```

```python
my_data.head(9500).plot(kind='scatter',x='Feature',y='Target')
```

```python
my_data.describe()
```

```python
my_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = my_data['Feature'].values.reshape(-1,1)
y = my_data['Target'].values.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_ #slope
```

```python
lm.intercept_
```

```python
# so our regression line is 
#y = lm.coef_ (x) + lm.intercept_
```

```python
predictions = lm.predict(x_test)
```
